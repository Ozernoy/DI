# üåü Exercise 2 : Creating a Line Plot for Temperature Variation
Instructions  
Create a simple line plot using Matplotlib that represents the temperature variations over a week.  
Use a list of temperature values for each day of the week (e.g., [72, 74, 76, 80, 82, 78, 75]) and label the x-axis as ‚ÄúDay‚Äù and the y-axis as ‚ÄúTemperature (¬∞F)‚Äù.  

```python
import matplotlib.pyplot as plt

# Temperature data for each day of the week
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperatures = [72, 74, 76, 80, 82, 78, 75]

# Creating the line plot
plt.plot(days, temperatures, marker='o')
plt.title('Temperature Variation Over a Week')
plt.xlabel('Day')
plt.ylabel('Temperature (¬∞F)')
plt.grid(True)
plt.show()
```

# üåü Exercise 3 : Visualizing Monthly Sales with a Bar Chart
Instructions  
Generate a bar chart using Matplotlib to visualize the monthly sales data for a retail store.  
Create a list of sales values for each month (e.g., [5000, 5500, 6200, 7000, 7500]) and label the x-axis as ‚ÄúMonth‚Äù and the y-axis as ‚ÄúSales Amount ($)‚Äù.  

```python
import matplotlib.pyplot as plt

# Monthly sales data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [5000, 5500, 6200, 7000, 7500]

# Creating the bar chart
plt.bar(months, sales, color='skyblue')
plt.title('Monthly Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.show()
```

# üåü Exercise 4 : Data Visualisation
Instructions  
Using the dataset sales_data.csv  

Calculate the total quantity of products sold by the company.  
Identify the category that had the highest revenue and calculate how much revenue it generated.  
Determine the average revenue per sale in the dataset.  
Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4).  

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the sales data
sales_data = pd.read_csv('https://raw.githubusercontent.com/intern2grow/sales-data-analysis/main/sales_data.csv')

# Total quantity of products sold
total_quantity = sales_data['quantity'].sum()
print(f'Total quantity of products sold: {total_quantity}')

# Category with the highest revenue
sales_data['revenue'] = sales_data['quantity'] * sales_data['price']
category_revenue = sales_data.groupby('category')['revenue'].sum()
highest_revenue_category = category_revenue.idxmax()
highest_revenue_value = category_revenue.max()
print(f'Category {highest_revenue_category} made {highest_revenue_value} revenue')

# Average revenue per sale
average_revenue_per_sale = sales_data['revenue'].mean()
print(f'{average_revenue_per_sale} is the average revenue per sale')

# Revenue generated in each quarter
sales_data['date'] = pd.to_datetime(sales_data['date'])
sales_data['quarter'] = sales_data['date'].dt.quarter
quarterly_revenue = sales_data.groupby('quarter')['revenue'].sum()
print(f'Revenue generated in each quarter: \n{quarterly_revenue}')

# Bar chart for total revenue in each quarter
plt.bar(quarterly_revenue.index, quarterly_revenue.values, color='purple')
plt.title('Total Revenue by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Total Revenue ($)')
plt.xticks([1, 2, 3, 4], ['Q1', 'Q2', 'Q3', 'Q4'])
plt.show()
```

# üåü Exercise 5 : Data Visualisation Using MatPlotLib
Instructions  
Create a simple line plot using Matplotlib. Plot the function y = x^2 for x values ranging from -10 to 10. Add a title and label the x and y axes.  
Generate a bar chart to display the sales data of four different products (A, B, C, D) in a week. The sales values are 15, 30, 45, and 20 respectively. Label the chart appropriately.  
Create a pie chart representing the following data about favorite fruits of a group of people: Apples (40%), Bananas (30%), Cherries (20%), and Dates (10%). Customize the chart with different colors for each fruit and add a legend.  

```python
# Line Plot

import matplotlib.pyplot as plt
import numpy as np

# x values and corresponding y values (y = x^2)
x = np.arange(-10, 11)
y = x**2

# Creating the line plot
plt.plot(x, y, marker='o')
plt.title('y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
# Bar Chart for product sales

products = ['A', 'B', 'C', 'D']
sales = [15, 30, 45, 20]

# Creating the bar chart
plt.bar(products, sales, color=['red', 'green', 'blue', 'orange'])
plt.title('Product Sales in a Week')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

```python
# Pie Chart for Favorite Fruits

fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [40, 30, 20, 10]
colors = ['red', 'yellow', 'pink', 'brown']

# Creating the pie chart
plt.pie(sizes, labels=fruits, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Favorite Fruits Distribution')
plt.legend(fruits, loc='best')
plt.show()
```
