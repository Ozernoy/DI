# Data loading

```python
import seaborn as sns
import pandas as pd

# Load the Iris dataset
iris = sns.load_dataset('iris')
```

# EDA

## The histogram for petal lengths

```python
import matplotlib.pyplot as plt

# Plot histogram of petal lengths
plt.figure(figsize=(8, 6))
plt.hist(iris['petal_length'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Petal Lengths in Iris Dataset')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

## The pair plot to show pairwise relationships in the dataset

```python
import seaborn as sns

# Create a pair plot
sns.pairplot(iris, hue='species', diag_kind='kde', height=2.5)
plt.suptitle('Pairwise Relationships in Iris Dataset', y=1.02)
plt.show()
```

## The scatter plot of sepal length vs. sepal width, color-coded by species

```python
from plotnine import ggplot, aes, geom_point, labs, theme_minimal

# Create scatter plot using Plotnine
plot = (ggplot(iris, aes(x='sepal_length', y='sepal_width', color='species'))
        + geom_point(size=2)
        + labs(title='Sepal Length vs. Sepal Width by Species',
               x='Sepal Length (cm)',
               y='Sepal Width (cm)')
        + theme_minimal())

plot.show()
```

## The interactive 3D scatter plot mapping sepal length, sepal width, and petal length, color-coded by species

```python
import plotly.express as px

# Create 3D scatter plot
fig = px.scatter_3d(iris, x='sepal_length', y='sepal_width', z='petal_length',
                    color='species',
                    title='3D Scatter Plot of Iris Dataset',
                    labels={
                        'sepal_length': 'Sepal Length (cm)',
                        'sepal_width': 'Sepal Width (cm)',
                        'petal_length': 'Petal Length (cm)'
                    })

fig.show()
```
