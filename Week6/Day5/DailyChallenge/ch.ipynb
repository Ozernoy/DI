# Statistical Analysis with NumPy and SciPy

```python
# Import necessary libraries
import numpy as np
from scipy import stats

# Extract numerical columns for statistical analysis
numerical_columns = mobile_data.select_dtypes(include=[np.number]).columns

# Calculate mean, median, and mode
central_tendency = {
    'Mean': mobile_data[numerical_columns].mean(),
    'Median': mobile_data[numerical_columns].median(),
    'Mode': mobile_data[numerical_columns].mode().iloc[0]
}

# Calculate range, variance, and standard deviation
variability = {
    'Range': mobile_data[numerical_columns].max() - mobile_data[numerical_columns].min(),
    'Variance': mobile_data[numerical_columns].var(),
    'Standard Deviation': mobile_data[numerical_columns].std()
}

# Calculate skewness and kurtosis
distribution_shape = {
    'Skewness': mobile_data[numerical_columns].skew(),
    'Kurtosis': mobile_data[numerical_columns].kurtosis()
}

# Combine all statistics into a single dataframe for better readability
statistics_summary = pd.DataFrame({
    'Mean': central_tendency['Mean'],
    'Median': central_tendency['Median'],
    'Mode': central_tendency['Mode'],
    'Range': variability['Range'],
    'Variance': variability['Variance'],
    'Standard Deviation': variability['Standard Deviation'],
    'Skewness': distribution_shape['Skewness'],
    'Kurtosis': distribution_shape['Kurtosis']
})

statistics_summary
```

## Hypothesis testing

- Null Hypothesis (H₀): The means of the given feature (e.g., battery power) are equal across all price range categories (0, 1, 2, and 3). In other words, there is no significant difference in the feature between the different price ranges.
- Alternative Hypothesis (H₁): At least one of the means of the feature (e.g., battery power) is different across the price range categories. In other words, the feature significantly varies between some or all price ranges.

```python
# Import the f_oneway function for ANOVA from SciPy
from scipy.stats import f_oneway

# Perform ANOVA on selected numerical features
anova_results = {}
features_to_test = ['battery_power', 'ram', 'px_height', 'px_width', 'talk_time']

for feature in features_to_test:
    # Perform ANOVA for each feature grouped by 'price_range'
    groups = [mobile_data[mobile_data['price_range'] == price][feature] for price in sorted(mobile_data['price_range'].unique())]
    anova_results[feature] = f_oneway(*groups)

# Convert the results to a more readable format
anova_summary = pd.DataFrame({
    'Feature': anova_results.keys(),
    'F-Statistic': [result.statistic for result in anova_results.values()],
    'P-Value': [result.pvalue for result in anova_results.values()]
})


anova_summary
```

- Battery Power, RAM, Pixel Height, and Pixel Width show statistically significant differences across the price ranges (p-values close to 0).
- Talk Time does not show a significant difference (p-value > 0.05), suggesting that it may not vary significantly across price ranges.

## Feature-target correlations

We can compute correlation coefficients between the numerical features and the target variable (price_range). We'll use Spearman's rank correlation and Pearson's correlation, two common methods for correlation analysis:

- Pearson's correlation: Measures the linear relationship between two variables.
- Spearman's rank correlation: Measures the monotonic relationship between two variables, making it more robust for non-linear relationships.

```python
# Import the necessary statistical functions
from scipy.stats import spearmanr, pearsonr

features_to_test = ['battery_power', 'ram', 'px_height', 'px_width', 'talk_time']

# Initialize dictionaries to store correlation results
pearson_corr = {}
spearman_corr = {}

# Calculate both Pearson and Spearman correlations between each feature and the target 'price_range'
for feature in features_to_test:
    pearson_corr[feature], _ = pearsonr(mobile_data[feature], mobile_data['price_range'])
    spearman_corr[feature], _ = spearmanr(mobile_data[feature], mobile_data['price_range'])

# Create a dataframe to summarize the correlation results
correlation_summary = pd.DataFrame({
    'Feature': features_to_test,
    'Pearson Correlation': pearson_corr.values(),
    'Spearman Correlation': spearman_corr.values()
})

correlation_summary
```

```python
# Import the necessary library for plotting
import matplotlib.pyplot as plt

features_to_test = ['battery_power', 'ram', 'px_height', 'px_width', 'talk_time']


# Create a bar plot for visualizing the Pearson and Spearman correlations
plt.figure(figsize=(10, 6))

# Bar width for two sets of correlations
bar_width = 0.35
index = range(len(features_to_test))

# Plot Pearson and Spearman correlations
plt.bar(index, pearson_corr.values(), bar_width, label='Pearson Correlation', color='blue')
plt.bar([i + bar_width for i in index], spearman_corr.values(), bar_width, label='Spearman Correlation', color='green')

# Add labels and title
plt.xlabel('Features')
plt.ylabel('Correlation with Price Range')
plt.title('Feature-Target Correlations (Pearson & Spearman)')
plt.xticks([i + bar_width / 2 for i in index], features_to_test, rotation=45)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

- RAM shows a very strong positive correlation with price_range, both in Pearson's (0.917) and Spearman's (0.917) correlations, indicating that higher RAM is strongly associated with higher price ranges.
- Battery Power, Pixel Height and Pixel Width show weak positive correlations with price_range
- Talk Time has very low correlation with price_range, indicating that it is not a significant determinant of the price.

## Advanced SciPy statistical functions
We'll apply Chi-Square Test to test the association between categorical features (e.g., blue, dual_sim) and the target variable (price_range).

```python
from scipy.stats import chi2_contingency

# List of boolean features for testing
categorical_features = ['blue', 'dual_sim', 'four_g', 'three_g', 'touch_screen', 'wifi']

# Dictionary to store chi-square results
chi2_results = {}

# Perform Chi-Square test for each categorical feature against 'price_range'
for feature in categorical_features:
    # Create a contingency table
    contingency_table = pd.crosstab(mobile_data[feature], mobile_data['price_range'])
    
    # Apply chi-square test
    chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)
    
    # Store the result
    chi2_results[feature] = {'Chi2 Statistic': chi2_stat, 'P-Value': p_val}

# Convert results to DataFrame for readability
chi2_summary = pd.DataFrame(chi2_results).T

chi2_summary
```

The Chi-Square test results for the categorical features show that none of these features have a statistically significant association with price_range (p-values > 0.05). This suggests that these features are not strongly related to the price range of phones.

# Vizualisation

```python
# Import necessary libraries for visualization
import seaborn as sns
import matplotlib.pyplot as plt

features_to_test = ['battery_power', 'ram', 'px_height', 'px_width', 'talk_time']


# Set the visual style
sns.set(style="whitegrid")

# 1. Histograms for numerical features
plt.figure(figsize=(12, 8))
for i, feature in enumerate(features_to_test):
    plt.subplot(2, 3, i+1)
    sns.histplot(mobile_data[feature], bins=20, kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 2. Scatter plots between selected features and the target 'price_range'
plt.figure(figsize=(12, 8))
for i, feature in enumerate(features_to_test):
    plt.subplot(2, 3, i+1)
    sns.scatterplot(x=mobile_data[feature], y=mobile_data['price_range'])
    plt.title(f'Scatter Plot: {feature} vs Price Range')
    plt.xlabel(feature)
    plt.ylabel('Price Range')
plt.tight_layout()
plt.show()

# 3. Box plots to show distribution of features across different price ranges
plt.figure(figsize=(12, 8))
for i, feature in enumerate(features_to_test):
    plt.subplot(2, 3, i+1)
    sns.boxplot(x='price_range', y=feature, data=mobile_data)
    plt.title(f'Box Plot: {feature} by Price Range')
    plt.xlabel('Price Range')
    plt.ylabel(feature)
plt.tight_layout()
plt.show()

# 4. Heatmap to show correlation matrix
plt.figure(figsize=(20, 8))
correlation_matrix = mobile_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

- Talk Time shows almost no correlation with the price range, indicating that this feature does not significantly affect the price classification of the phones.

- However, features like Talk Time do not show significant differences across price ranges, suggesting that they are less important for price classification.

- The heatmap of correlations confirmed that RAM is the feature most strongly correlated with price. Other features like battery power, pixel height, and pixel width also show moderate correlations.

These features might be important for certain consumers but do not seem to play a significant role in determining the price range in this dataset.
