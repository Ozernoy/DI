# ðŸŒŸ Exercise 2 : Descriptive Statistics
Instructions
Given a sample dataset, calculate the mean, median, variance, and standard deviation using SciPy.  
Sample dataset in code:
```python
data = [12, 15, 13, 12, 18, 20, 22, 21]
```

````python
import numpy as np
from scipy import stats

data = [12, 15, 13, 12, 18, 20, 22, 21]


scipy_stats = stats.describe(data)


mean = scipy_stats.mean
variance = scipy_stats.variance
std_dev = np.sqrt(variance)  # Standard deviation is the square root of variance
median = np.median(data)

mean, median, variance, std_dev
````

# ðŸŒŸ Exercise 3 : Understanding Distributions
Instructions  
Generate a normal distribution using SciPy with a mean of 50 and a standard deviation of 10. Plot this distribution.

````python
import matplotlib.pyplot as plt

mean, std_dev = 50, 10

data_normal = np.random.normal(mean, std_dev, 1000)

plt.figure(figsize=(8, 6))
plt.hist(data_normal, bins=30, density=True, alpha=0.6, color='b', edgecolor='black')
plt.title('Normal Distribution with Mean = 50 and Std Dev = 10')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
````

# ðŸŒŸ Exercise 4 : T-Test Application
Instructions
Perform a T-test on two sets of randomly generated data. Hint: Use numpy.random.normal() to generate the datasets.

````python
from scipy import stats

data1 = np.random.normal(50, 10, 100)  # Mean = 50, Std = 10, 100 samples
data2 = np.random.normal(55, 12, 100)  # Mean = 55, Std = 12, 100 samples

t_stat, p_value = stats.ttest_ind(data1, data2)

t_stat, p_value
````
