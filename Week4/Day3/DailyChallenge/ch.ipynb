```python
import zipfile
import os

# Define the path to your ZIP file and the extraction directory
zip_file_path = 'flower-color-images.zip'
extract_dir = 'flower-color-images'

# Create the extraction directory if it doesn't exist
if not os.path.exists(extract_dir):
    os.makedirs(extract_dir)

# Extract the ZIP file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print("Extraction completed!")
```

```python
import os
import matplotlib.pyplot as plt
from PIL import Image, ImageOps

# Function to display images
def show_images(original, augmented, title):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(original)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(augmented)
    plt.title(title)
    plt.axis('off')

    plt.show()

# Load an example image from the dataset
image_path = 'flower-color-images/flower_images/flower_images/0001.png'
original_image = Image.open(image_path)

# 1. Rotate the image by 90 degrees
rotated_image = original_image.rotate(90)

# 2. Flip the image horizontally and then vertically
flipped_image = ImageOps.mirror(original_image)
flipped_image = ImageOps.flip(flipped_image)

# 3. Zoom in on the image (scale by 1.2x)
width, height = original_image.size
zoomed_image = original_image.resize((int(width * 1.2), int(height * 1.2)))

# Display the images
show_images(original_image, rotated_image, 'Rotated Image (90 degrees)')
show_images(original_image, flipped_image, 'Flipped Image (Horizontally and Vertically)')
show_images(original_image, zoomed_image, 'Zoomed Image (1.2x)')
```
