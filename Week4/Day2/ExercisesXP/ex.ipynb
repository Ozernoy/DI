# Mental Health Data

Dataset contains data on percent of several mental health disorders observed in 276 countries or regions from 1776 till 2019.

**Columns categories**  
quantitative: Year, all disorder columns  
qualitative:  indes, Entity, Code

Explanation: all columns to which summing, finding mean and doing other mathematical operations is reasonable can be considered quantitive, columns which describe or label some attribute - qualitive

```python
mental_health_data.head()
```

```python
mh_stats = mental_health_data.describe()
print(mh_stats.columns)
mh_quantitative_cols = ['Anxiety disorders (%)', 'Drug use disorders (%)',
       'Depression (%)', 'Alcohol use disorders (%)']
mh_stats = mh_stats[mh_quantitative_cols]
mh_stats
```

```python
scatter_matrix(mental_health_data[mh_quantitative_cols], figsize=(12,15), diagonal='hist')
plt.show()
```

```python
mh_corr = mental_health_data[mh_quantitative_cols].corr()
mh_corr
```

```python
sns.heatmap(mh_corr, annot=True, cmap='coolwarm', fmt=".2f")
```

**Insights and conclusion**  
Anxiety disorders are seen in a large percentage of the population, followed by depression, alcohol and drugs  
There is a positive correlation (0.62) between the presence of anxiety disorders and drug use

# Sleep Data
The dataset provides information on the average amount of time Americans spend sleeping, broken down by age group, sex, and activity

**Columns categories**  
quantitative: Year, Avg hrs per day sleeping, Standard Error  
qualitative:  index, Perios, Type of Days, Age Group, Activity, Sex  


```python
sleep_data.head()
```

```python
s_stats = sleep_data.describe()
print(s_stats.columns)
s_quantitative_cols = ['Year', 'Avg hrs per day sleeping', 'Standard Error']
s_stats[s_quantitative_cols]
```

```python
sleep_data[s_quantitative_cols].hist(bins=10, figsize=(10, 8))
plt.show()
```

```python
s_corr = sleep_data[s_quantitative_cols].corr()
s_corr
```

```python
sns.heatmap(s_corr, annot=True, cmap='coolwarm', fmt=".2f")
```

**Insights and conclusion**  
On average, Americans sleep 8.8 hours.   
The maximum recorded average sleep time in a group is 10.65 (this is not the same as just the maximum sleep time, since each entry in the table is already a group, not an individual)
