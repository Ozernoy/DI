# ðŸŒŸ Exercise 3 : Scrape Dynamic Content from Rotten Tomatoes
Task:
Use Selenium to navigate to the Rotten Tomatoes Certified Fresh Movies page.  
Extract the HTML content after itâ€™s fully loaded.  
Use BeautifulSoup to parse and extract the movie titles, scores, and release dates.  
Instructions  
Set up Selenium WebDriver and navigate to the Rotten Tomatoes page.  
Extract the HTML content using driver.page_source.  
Parse the HTML with BeautifulSoup.  
Find and extract the desired movie information.  
Print the extracted data.  

````python
from bs4 import BeautifulSoup
from selenium import webdriver
import pprint  # To tidy up the output

# Set up Chrome WebDriver with options
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Run Chrome in headless mode
options.add_argument("--no-sandbox")  # Bypass OS security model
options.add_argument("--disable-dev-shm-usage")  # Overcome limited resource problems

driver = webdriver.Chrome(options=options)

# Navigate to the Rotten Tomatoes page
url = "https://www.rottentomatoes.com/browse/movies_at_home/affiliates:netflix~critics:certified_fresh"
driver.get(url)

# Extract the page source and parse it with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all movie elements using the main container class 'flex-container'
movies = soup.find_all('div', class_='flex-container')

# Loop through each movie and extract title, score, and release date
for movie in movies:
    # Extract title
    title_element = movie.find('span', class_='p--small', attrs={"data-qa": "discovery-media-list-item-title"})
    title = title_element.get_text(strip=True) if title_element else "No title"

    # Extract release date
    release_date_element = movie.find('span', class_='smaller', attrs={"data-qa": "discovery-media-list-item-start-date"})
    release_date = release_date_element.get_text(strip=True) if release_date_element else "No release date"

    # Extract critics score
    score_element = movie.find('rt-text', attrs={"slot": "criticsScore"})
    score = score_element.get_text(strip=True) if score_element else "No score"

    # Print the extracted data
    print(f"Title: {title}")
    print(f"Score: {score}")
    print(f"Release Date: {release_date}")
    print("-------------")

# Close the WebDriver
driver.quit()
````

# ðŸŒŸ Exercise 4 : Scrape and Categorize News Articles from a JavaScript-Enabled News Site
Task:  
Visit a section of a news website (e.g., the Technology section of BBC News).  
Scrape news article titles and their publication dates.  
Categorize articles based on their publication month.  
Instructions:  
Use Selenium to navigate to a specific news section on the website.  
Extract and parse the HTML content that is dynamically loaded via JavaScript.  
Using BeautifulSoup, extract news article titles and publication dates.  
Categorize articles by their publication month (e.g., â€˜Januaryâ€™, â€˜Februaryâ€™, etc.).  
Print the categorized lists of articles.  

Let's webscrap reddit

````python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
import time

# Set up Selenium WebDriver with custom user-agent
options = webdriver.ChromeOptions()
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36")
options.add_argument('--headless')  # Optional: run browser in headless mode
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Initialize the WebDriver with the options
driver = webdriver.Chrome(options=options)

url = 'https://www.reddit.com/'  
driver.get(url)

# Adjustable number of scrolls
SCROLL_LIMIT = 3  

# Function to scroll down the page
def scroll_page():
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)  # Wait for new content to load after each scroll

# Perform scrolling a set number of times
for scroll_count in range(SCROLL_LIMIT):
    print(f"Scrolling {scroll_count + 1} of {SCROLL_LIMIT}...")
    scroll_page()

# After scrolling, get the page source and parse it with BeautifulSoup
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# Lists to store titles and dates
titles = []
dates = []

# Step 1: Extracting all post titles (headings)
headings = soup.find_all('a', attrs={'slot': 'title', 'class': 'block font-semibold text-neutral-content-strong m-0 visited:text-neutral-content-weak text-16 xs:text-18 mb-2xs xs:mb-xs'})
for heading in headings:
    title_text = heading.get_text(strip=True)
    titles.append(title_text)

# Step 2: Extracting post dates
date_elements = soup.find_all('time')
for date in date_elements:
    datetime_value = date.get('datetime', 'No datetime found')
    dates.append(datetime_value)

# Close the WebDriver
driver.quit()

# Step 3: Create a DataFrame with titles and dates
df = pd.DataFrame({
    'Title': titles,
    'Date': pd.to_datetime(dates)  # Convert dates to datetime format for easier grouping
})

df
````

````python
'''
Visit a specific cityâ€™s weather forecast page on a weather website.  
Scrape weather forecast data including temperature, condition, and humidity.  
Analyze the data to find the average temperature and most common weather condition.  
Instructions:  
Use Selenium to navigate to the weather forecast page of a specific city.    
Extract and parse the HTML content, focusing on dynamically loaded weather data.  
Using BeautifulSoup, extract relevant weather information like temperature, condition (sunny, cloudy, etc.), and humidity.  
Calculate the average temperature and identify the most common weather condition.  
Print the analysis results.  
'''


from turtle import st
from numpy import disp
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
import re
from datetime import datetime
import pandas as pd

# Initialize lists to store data
date = []
h_temp = []
l_temp = []
rf_temp = []
rf_shade_temp = []
max_uv_index = []
wind = []
humidity = []

df = pd.DataFrame()

# Set Chrome options
options = webdriver.ChromeOptions()
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36")
# options.add_argument('--headless')  # Run browser in headless mode
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--window-position=-10000,0')  # Move window off-screen

# Initialize WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Open the URL
url = 'https://www.accuweather.com/en/ru/moscow/294021/daily-weather-forecast/294021'
driver.get(url)

# Scroll the whole page
last_height = driver.execute_script("return document.body.scrollHeight")

n_scrolls = 1
while True:
    print(f'Making scroll {n_scrolls}')
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)  # Wait for new content to load
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
    n_scrolls += 1

# Get the page source and parse it with BeautifulSoup
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# Extract data and add it to the lists

date_parse = soup.find_all('span', class_='module-header sub date')
for d in date_parse:
    date.append(d.get_text(strip=True))

h_temp_parse = soup.find_all('span', class_='high')
for h in h_temp_parse:
    h_temp.append(int(h.get_text(strip=True).replace('Â°', '')))

l_temp_parse = soup.find_all('span', class_='low')
for l in l_temp_parse:
    l_temp.append(int(l.get_text(strip=True).replace('Â°', '').replace('/', '')))

humidity_parse = soup.find_all('div', class_='precip')
for hum in humidity_parse:
    humidity.append(int(hum.get_text(strip=True).replace('%', '')))

# Parsing panel_items to extract RealFeel temperature, RealFeel Shade temperature, Max UV Index and Wind Speed
panel_items = soup.find_all('p', class_='panel-item')
for item in panel_items:

    # Check if the text 'RealFeelÂ®' is in the <p> tag
    if 'RealFeelÂ®' in item.get_text():
        # Find the nested <span> tag with the class 'value'
        value_span = item.find('span', class_='value')
        if value_span:
            # Extract and clean the text content to get the number
            value_text = value_span.get_text(strip=True)
            # Remove the degree symbol and convert to integer
            value_number = int(value_text.replace('Â°', ''))
            rf_temp.append(value_number)
    
    # Check if the text 'RealFeel Shade' is in the <p> tag
    if 'RealFeel Shadeâ„¢' in item.get_text():
        value_span = item.find('span', class_='value')
        if value_span:
            value_text = value_span.get_text(strip=True)
            value_number = int(value_text.replace('Â°', ''))
            rf_shade_temp.append(value_number)
    
    if 'Max UV Index' in item.get_text():
        value_span = item.find('span', class_='value')
        if value_span:
            value_text = value_span.get_text(strip=True)
            value_number = int(re.search(r'\d+', value_text).group())
            max_uv_index.append(value_number)
    
    if 'Wind' in item.get_text():
        value_span = item.find('span', class_='value')
        if value_span:
            value_text = value_span.get_text(strip=True)
            value_number = int(re.search(r'\d+', value_text).group())
            wind.append(value_number)

# Create a DataFrame with the extracted data

# Convert 'MM/DD' format to a proper date format, assuming the current year
current_year = datetime.now().year
formatted_dates = [datetime.strptime(f"{current_year}/{d}", "%Y/%m/%d").strftime("%Y-%m-%d") for d in date]

df = pd.DataFrame({
    'Date': formatted_dates,
    'High Temp': h_temp,
    'Low Temp': l_temp,
    'RealFeel Temp': rf_temp,
    'RealFeel Shade Temp': rf_shade_temp,
    'Max UV Index': max_uv_index,
    'Wind Speed': wind,
    'Humidity': humidity
})

display(df.head())

# Close the driver
driver.quit()
````
